### Дано: f(x) = 0.6x^3 + 5.5x^2 + 10x - 5

1. Определить корни
2. Найти интервалы, на которых функция возрастает
3. Найти интервалы, на которых функция убывает
4. Построить график
5. Вычислить вершину
6. Определить промежутки, на котором f > 0
7. Определить промежутки, на котором f < 0

```python
from sympy import *
a = 0.6
b = 5.5
c = 10
d = -5

x = Symbol('x')
f_x = a * (x**3) + b * (x**2) + c * x + d
```

```python
f_x
```

### 1. Корни функции

```python
roots = solve(f_x,x)
for i in range(1, len(roots) + 1):
    print(f'Корень №{i} = {roots[i - 1].n(2)}')
```

```python
derivative = diff(f_x)
derivative
```

### 2. Интервалы, на которых функция возрастает

```python
solve(derivative > 0, x)
```

### 3. Интервалы, на которых функция убывает

```python
solve(derivative < 0, x)
```

### 4. График функции

```python
from sympy.plotting import plot

g3 = plot(f_x, derivative, (x, -7, 3), title='Совмещенный график', legend=True, xlabel='X', ylabel='Y', show=False)
g3[0].line_color = "red"
g3[1].line_color = "blue"
g3.show()
```

### 5. Вершины функции

```python
roots_dev = solve(derivative,x)
for i in range(1, len(roots_dev) + 1):
    print(f'Корень производной №{i} = {roots_dev[i - 1].n(2)}')
```

### 6. Промежутки, на котором f > 0

```python
solve(f_x > 0, x)
```

### 7. Промежутки, на котором f < 0

```python
solve(f_x < 0, x)
```
